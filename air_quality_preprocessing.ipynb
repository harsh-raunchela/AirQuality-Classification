```python
# #now we need to calculate the aqi (indian standard)
# AQI Range	Category
# 0–50	Good
# 51–100	Satisfactory
# 101–200	Moderate
# 201–300	Poor
# 301–400	Very Poor
# 401–500	Severe
```

```python
def get_aqi_category_pm25(pm):
    if pm <= 30:
        return "Good"
    elif pm <= 60:
        return "Satisfactory"
    elif pm <= 90:
        return "Moderate"
    elif pm <= 120:
        return "Poor"
    elif pm <= 250:
        return "Very Poor"
    else:
        return "Severe"
```

```python
df['AQI_Category'] = df['PM2.5 (µg/m³)'].apply(get_aqi_category_pm25)
```

```python
df['AQI_Category'].value_counts()
```

```python
#fix the timestamp column 
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df['Month'] = df['Timestamp'].dt.month
df['Hour'] = df['Timestamp'].dt.hour
df['Weekday'] = df['Timestamp'].dt.weekday

df.drop(columns=['Timestamp'], inplace=True)
```

```python
cols_to_drop = ['PM10 (µg/m³)', 'NO (µg/m³)','TOT-RF (mm)']
X_selected = X_scaled_df.drop(columns = cols_to_drop)
X_selected.shape
```

```python
# #lets do selectKbest now
# It uses a statistical test to score each feature based on how strongly it relates to the target (AQI_Label). The higher the score, the more useful the feature is.
# For classification problems like ours, we use f_classif (ANOVA F-test) as the scoring function.
# It then selects the top K features based on those scores.
#for applly this i will ask sir if we can apply this or not till then..bye bye 
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

X_features = X_selected.drop(columns=['AQI_Label'])
y_labels = X_selected['AQI_Label']

selector = SelectKBest(f_classif, k=15)
selector.fit(X_features, y_labels)

feature_scores = pd.DataFrame({
    'Feature': X_features.columns,
    'Score': selector.scores_
})

feature_scores = feature_scores.sort_values('Score', ascending=False)
print(feature_scores)
```

```python
cols_to_drop_2 = ['Weekday','Hour', 'RF (mm)']
X_final = X_selected.drop(columns=cols_to_drop_2)
print(X_final.shape)
```
