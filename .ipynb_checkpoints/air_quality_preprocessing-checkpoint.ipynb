```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("data.csv")

# See first 5 rows
df.head()
```

```python
#as the whole column is empty 
df.drop(columns=['O Xylene (µg/m³)'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# #now we need to calculate the aqi (indian standard)
# AQI Range	Category
# 0–50	Good
# 51–100	Satisfactory
# 101–200	Moderate
# 201–300	Poor
# 301–400	Very Poor
# 401–500	Severe
```

```python
def get_aqi_category_pm25(pm):
    if pm <= 30:
        return "Good"
    elif pm <= 60:
        return "Satisfactory"
    elif pm <= 90:
        return "Moderate"
    elif pm <= 120:
        return "Poor"
    elif pm <= 250:
        return "Very Poor"
    else:
        return "Severe"
```

```python
df['AQI_Category'] = df['PM2.5 (µg/m³)'].apply(get_aqi_category_pm25)
```

```python
df['AQI_Category'].value_counts()
```

```python
#fix the timestamp column 
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df['Month'] = df['Timestamp'].dt.month
df['Hour'] = df['Timestamp'].dt.hour
df['Weekday'] = df['Timestamp'].dt.weekday

df.drop(columns=['Timestamp'], inplace=True)
```

```python
#creates aqi category to numeric 

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Create numeric target separately
y = le.fit_transform(df['AQI_Category'])
```

```python
X = df.drop(columns=['AQI_Category'])
```

```python
#apply standard scalling for standardization 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#convert it back to dataframe 


X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df['AQI_Label'] = y
```
