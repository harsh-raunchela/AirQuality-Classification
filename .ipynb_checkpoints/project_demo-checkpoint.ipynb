```python
import joblib
import pandas as pd
```

```python
model = joblib.load("svm_model.pkl")
scaler = joblib.load("scaler.pkl")
le = joblib.load("label_encoder.pkl")
```

```python
df = pd.read_csv("final_data.csv")
X = df.drop(columns=['AQI_Label'])
```

```python
df
```

```python
X
```

```python
pred = model.predict(X.iloc[[0]])

print("Predicted Label:", pred[0])
print("Predicted Category:", le.inverse_transform(pred)[0])
```

```python
row = 100

pred = model.predict(X.iloc[[row]])

print("Sample Row:", row)
print("Predicted Label:", pred[0])
print("Predicted Category:", le.inverse_transform(pred)[0])
print("Actual Label:", df['AQI_Label'].iloc[row])
```

```python
for i in range(5):
    pred = model.predict(X.iloc[[i]])
    print(i, le.inverse_transform(pred)[0])
```

```python
row = int(input("Enter row number (0 to 8783): "))

pred = model.predict(X.iloc[[row]])

print("Predicted AQI:", le.inverse_transform(pred)[0])
print("Actual AQI Label:", le.inverse_transform([df['AQI_Label'].iloc[row]])[0])
```
