```python
#clean column names 
poll_df.columns = poll_df.columns.str.replace(' ', '_')
poll_df.columns = poll_df.columns.str.replace('(', '')
poll_df.columns = poll_df.columns.str.replace(')', '')
poll_df.columns = poll_df.columns.str.replace('/', '')
poll_df.columns = poll_df.columns.str.replace('µg/m³', 'ugm3')
poll_df.columns = poll_df.columns.str.replace('°C', 'C')
poll_df.columns = poll_df.columns.str.replace('%', 'percent')
poll_df.columns = poll_df.columns.str.replace('-', '_')

print(poll_df.columns)
```

```python
#now we have to convert the timestamp column to datetime format 
poll_df['Timestamp'] = pd.to_datetime(poll_df['Timestamp'])
poll_df['Timestamp'].head()
```

```python
#in our dataset there is no AQI column so we have to calculate teh aqi then make a category for aqi 
# Indian AQI standard:

# 0–50 → Good

# 51–100 → Satisfactory

# 101–200 → Moderate

# 201–300 → Poor

# 301–400 → Very Poor

# 401+ → Severe

# We’ll apply this on PM2.5.
```

```python
poll_df.shape
```

```python
#since we have hourly data we have to convert it to daily average bcox aqi is calculated like that

#first the timestamp data in date column
poll_df['Date'] = poll_df['Timestamp'].dt.date

#now we have to group by date and then we will take the daily average
daily_df = poll_df.groupby('Date').mean(numeric_only=True).reset_index()

#now convert date back to date time 
daily_df['Date'] = pd.to_datetime(daily_df['Date'])

daily_df.shape
```

```python
daily_df.head()

#from this we have get that we have 25 features, daily averaged data, date column, pollutants and weather , many nan values 
```

```python
#next step is to clean the daily dataset propely as ther are nan values

#check missing value percentage if a coulumn have 70% missing value we will droop that column

miss_percentage = daily_df.isnull().mean() * 100
miss_percentage.sort_values(ascending = False)
```

```python
#therer are many columns with 100 % missing now we will drop them 

cols_to_drop = [
    'NH3_µgm³',
    'RF_mm',
    'TOT_RF_mm',
    'Eth_Benzene_µgm³',
    'MP_Xylene_µgm³',
    'O_Xylene_µgm³',
    'Xylene_µgm³',
    'Toluene_µgm³',
    'Benzene_µgm³',
    'VWS_ms'
]

daily_df = daily_df.drop(columns=cols_to_drop)

daily_df.shape
```

```python
daily_df.head()
```

```python
#now we will make a aqi category we have set  a range for each category according to indian aqi category 

def categorize_aqi(pm25):
    if pm25 <= 50:
        return "Good"
    elif pm25 <= 100:
        return "Satisfactory"
    elif pm25 <= 200:
        return "Moderate"
    elif pm25 <= 300:
        return "Poor"
    elif pm25 <= 400:
        return "Very Poor"
    else:
        return "Severe"

daily_df['AQI_Category'] = daily_df['PM2.5_µgm³'].apply(categorize_aqi)

#now we have make a aqi category lets check what values are given 

daily_df[['PM2.5_µgm³', 'AQI_Category']].head()
```
